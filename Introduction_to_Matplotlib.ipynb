```python
import matplotlib.pyplot as plt
import matplotlib.pyplot as plt

student_name = [
    "Ava Sinclair",
    "Liam Torres",
    "Nia Bennett",
    "Marcus Reed",
    "Zoe Caldwell",
    "Jasper Lin",
    "Elena Cruz",
    "Kai Thompson",
    "Ruby Patel",
    "Theo Morgan"
]
student_marks = [45, 38, 42, 47, 35, 40, 44, 36, 41, 39]
for name,marks in zip (student_name,student_marks):
    print(f"{name}: {marks} marks")
```

```python
def line_chart ():
  plt.plot(student_name, student_marks, color='skyblue')
  plt.title("Line Chart of Test Marks")
  plt.xlabel("Student Names")
  plt.ylabel("Student Marks")
  plt.xticks (rotation = 45)
  plt.ylim(0,50)
  plt.grid (True)
  plt.show()
line_chart ()
```
