```python
import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt
```

```python
df = pd.read_csv("Iris Dataset.csv")
df.head()
```

```python
plt.figure(figsize=(8, 6))
sns.barplot(data=df, x='Species', y='SepalLengthCm', palette="Paired")
plt.title("Sepal Length vs Species")
plt.show()
```

```python
cubehelix_colors = sns.cubehelix_palette(start=0.5, rot=-0.5, n_colors=3)
plt.figure(figsize=(6, 5))
sns.countplot(data=df, x='Species', palette=cubehelix_colors)
plt.title("Count of Each Species")
plt.show()
```

```python
plt.figure(figsize=(8, 6))
icefire = sns.color_palette("icefire", n_colors=3)
sns.boxplot(data=df, x='Species', y='SepalWidthCm', palette=icefire)
plt.title("Boxplot of Sepal Width by Species")
plt.show()
```

```python
plt.figure(figsize=(8, 6))
sns.swarmplot(data=df, x='Species', y='SepalWidthCm', palette="husl")
plt.title("Swarm Plot of Sepal Width by Species")
plt.show()
```

```python
sns.displot(df['SepalWidthCm'], kde=True, color='navy')
plt.title("Distribution of Sepal Width")
plt.show()
```

```python
sns.jointplot(data=df, x='SepalLengthCm', y='SepalWidthCm', kind='scatter',
              color='#E76F51')
plt.suptitle("Joint Plot of Sepal Length vs Sepal Width", y=1.02)
plt.show()
```

```python
sns.pairplot(df, hue='Species', diag_kind='kde', palette='Set2')
plt.suptitle("Pair Plot of All Features with 'Set2' Palette", y=1.02)
plt.show()
```
