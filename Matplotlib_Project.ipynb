```python
import matplotlib.pyplot as plt
fruits = [
    'Apples', 'Bananas', 'Cherries', 'Grapes', 'Oranges',
    'Mangoes', 'Strawberries', 'Pineapples', 'Kiwis', 'Blueberries'
]
sales = [150.5, 200.2, 120.75, 180.6, 170.1, 90.45, 140.3, 110.8, 80.95, 130.4]
colors = [
    '#FF0000', '#FFFF00', '#8B0000', '#6A0DAD', '#FFA500',
    '#FFB347', '#FF69B4', '#FFD700', '#9ACD32', '#0000FF'
]
```

```python
plt.figure(figsize=(10, 6))
bars = plt.bar(fruits, sales, color=colors, edgecolor='black', linewidth=1.7)
plt.grid(axis='y', linestyle='--', alpha=0.2)
plt.title('Weekly Fruit Sales (in Units)', fontsize=24, fontfamily='sans-serif', color='darkred')
plt.xlabel('Fruits', fontsize=16, fontfamily='serif', color='darkgreen')
plt.ylabel('Units Sold', fontsize=16, fontfamily='serif', color='darkgreen')
plt.xticks(rotation=45)
plt.tight_layout()
plt.show()
```
