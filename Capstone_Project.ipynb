**Step 2: Inserting the neccessary Python Libraries**
This is part to import the 4 important libraries in the progamming language "Python" for Data Visualization.

```python
import pandas as pd
import matplotlib.pyplot as plt
import numpy as np
import seaborn as sns
```

**Step 3: Reading the file and displaying 21 fields of the data set**
In this step, you'll read the first 21 data inside the file.

```python
df = pd.read_csv("penguins_size.csv")
df.head(21)
```

**Step 4: Removal of the null values found in the data**

```python
df.isnull().sum()
```

```python
df.info()
```

**Step 5: Statistics Summary**

```python
df.describe()
```

**Step 6: Drawing a Count Plot**

This is done by using Seaborn libary

```python
custom_palette = ["#FF5733", "#33FF57", "#3357FF"]
sns.countplot(data=df, x="species", hue="island", palette=custom_palette)
plt.show()
```

**Step 7: Compare Feautures by Box-Whisker Plot**

```python
df.plot(
    kind='box',
    subplots=True,
    layout=(3, 2),
    sharex=False,
    sharey=False,
    figsize=(8, 12))
plt.show()
```

**Step 8: Sketching a Histogram**

```python
df.hist(figsize=(12, 8), color='skyblue')
plt.show()
```

**Step 8:Sketching a HeatMap**

```python
numerical_df = df.select_dtypes(include=['number'])
sns.heatmap(numerical_df.corr(),cmap= "cividis", annot = True)
```
