```python
import seaborn as sns
import pandas as pd
import numpy as py
import matplotlib.pyplot as plt
```

```python
df = pd.read_csv('Tips Dataset.csv')
```

```python
df.head(5)
```

```python
df.info()
```

```python
df.describe()
```

```python
df[['size', 'tip', 'total_bill']].hist(bins=15, figsize=(10, 5))
```

```python
sns.histplot(data=df, x='total_bill', kde=True, color='skyblue')
plt.title("Histogram of Total Bill with KDE")
plt.xlabel("Total Bill ($)")
plt.ylabel("Frequency")
plt.show()
```

```python
sns.scatterplot(data=df, x='total_bill', y='tip', hue='time', palette='coolwarm')
plt.title("Total Bill vs Tip by Time")
plt.xlabel("Total Bill ($)")
plt.ylabel("Tip ($)")
plt.show()
```

```python
sns.pairplot(data=df, hue='sex', palette='husl')
plt.suptitle("Pair Plot with Gender as Hue", y=1.02)
plt.show()
```
